```python
y = np.array([0,3,6,9,13,16])
plt.plot(y,marker='>')  
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("The marked graph")        # Used to mark the points in graph
plt.show()
```

```python
y = np.array([0,3,6,9,13,16])
plt.plot(y,'*-.b')              #if X-axis is not given 
# -- , : , -. , -  are some line patterns
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("The marked graph")        # Used to mark the points in graph
plt.show()
```

```python

y = np.array([1, 3, 5, 7, 13, 19, 10, 23])

# Create subplot 1
plt.subplot(1, 2, 1)  # 1 row, 2 columns, subplot 1
plt.plot(y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("The Plot1 graph")
plt.grid()

y1 = np.array([0, 3, 6, 9, 13, 16])

# Create subplot 2
plt.subplot(1, 2, 2)  # 1 row, 2 columns, subplot 2
plt.plot(y1)
plt.grid()
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("The Plot2 graph")

plt.suptitle("The plot graph")          # Super title for the graph

plt.show()
```
