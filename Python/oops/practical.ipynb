```python
# Sample from iris species
species = ['set0' , 'set0', 'set0', 'set0',
           'color1', 'color1', 'color1', 'color1', 'color1', 'color1',
           'black2', 'black2', 'black2', 'black2', 'black2', 'black2', 'black2']

# Create a dictionary to map species to group IDs
species_to_id = {'set0': 0, 'color1': 1, 'black2': 2}

# Use list comprehension to map species to group IDs
group_ids = [species_to_id[spec] for spec in species]

print(group_ids)
```

```python

df_filtered = df[['Survived', 'Pclass']]
survival_rate = df_filtered.groupby('Pclass')['Survived'].mean()

survival_rate.plot(kind='bar', color='skyblue')

# Adding labels and title
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Passenger Class')

# Showing the plot
plt.show()
```

```python

df_filt = df[['Survived', 'Sex']]
survival_rate = df_filt.groupby('Sex')['Survived'].mean()

survival_rate.plot(kind='bar', color='skyblue')

# Adding labels and title
plt.xlabel('Passenger Gender')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Passenger Class')

# Showing the plot
plt.show()
```

```python
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris().head(16)
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target

# Create group IDs based on the 'species' column
group_ids = df.groupby('species').cumcount()

# Convert to list
group_ids_list = group_ids.tolist()

# Print the desired output
print(group_ids_list)
```
