```python
import pandas as pd
import matplotlib.pyplot as plt

# Example DataFrame
data = {'Gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
        'Age': [25, 30, 35, 28, 32]}
df1 = pd.DataFrame(data)

# Group by 'Gender' and calculate the mean age for each group
grouped_data = df1.groupby('Gender')['Age'].mean()  # Applying mean() to calculate the mean age
print(grouped_data)

# Plot the graph
grouped_data.plot(kind='bar', color=['blue', 'pink'])  # Bar plot with different colors for each group
plt.xlabel('Gender')
plt.ylabel('Mean Age')
plt.title('Mean Age by Gender')
plt.xticks(rotation=0)  # Rotate x-axis labels if needed
plt.show()
```
